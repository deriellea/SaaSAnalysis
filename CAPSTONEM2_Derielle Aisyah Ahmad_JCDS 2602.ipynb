# Latar Belakang
AWS atau Amazon Web Service merupakan anak perusahaan dari Amazon yang menyediakan platform *cloud-computing* dan API sesuai permintaan clientnya dengan sistem pembayaran berkala. Terdapat 3 model *cloud-computing* yang ditawarkan oleh AWS yaitu *Infrastructure as a Service*(IaaS), *Platform as a Service* (PaaS), dan *Software as a Service* (SaaS). 

*Software as as Service* adalah model penyampaian perangkat lunak (software) melalui internet. Aplikasi SaaS ditujukan untuk digunakan oleh banyak penguna dan merupakan aplikasi yang ditanamkan pada *cloud infrastructure* [Fathoni dan Kenali, 2014](https://media.neliti.com/media/publications/102046-ID-desain-dan-implementasi-software-as-serv.pdf). Pengguna SaaS umumnya dapat mengakses aplikasi secara online melalui *cloud service provider* seperti  internet. Contoh dari *Software as a Service* adalah seperi program dari Adobe, Google, AWS, dan lain-lain. 

AWS mendefinisikan SaaS sebagai model pengiriman bisnis dan perangkat luank yang memungkinkan organisasi untuk menyediakan penawaran/produk mereka kepada pelanggan denga cara yang berpusat dan minim hambatan ([AWS](https://aws.amazon.com/saas/#:~:text=AWS%20approaches%20Software%2Das%2Da,friction%20and%20service%2Dcentric%20manner.)). Dari pernyataan tersebut dapat diketahui bahwa AWS menyediakan produk SaaS untuk perusahaan lain (B2B). Dimana AWS bekerja sama dengan organisasi untuk membangun model SaaS yang sesuai dengan kebutuhan organisasi sehingga didapat kelincahan dan efisiensi operasional bisnis untuk mendorong pertumbuhan, jangkauan, dan inovasi. AWS sebagai salah satu pemain yang dominan pada pasar global *cloud infrasturcutre* merupakan penyedia layanan *cloud* terdepan pada tahun 2020, menyumbang 31% dari total pengeluaran ([Catalyst, 2021](https://www.canalys.com/newsroom/global-cloud-market-q4-2020)). Di tahun 2022, AWS menjadi segmen yang paling cepat berkembang dan menguntungkan bagi Amazon. Di tahun ini, Amazon menghasilkan pendapatan sebesar $80 miliar dari layanan AWS. Selain itu, pangsa pasarnya terhadap total penjualan bersih terus meningkat, naik dari 12% pada tahun 2020 menjadi 16% pada tahun 2022. Dengan margin laba operasi sebesar 29%, AWS juga merupakan segmen yang paling menguntungkan bagi perusahaan ([Forrester, 2023](https://www.forrester.com/blogs/amazon-sales-and-profit-analysis-for-2022-top-10-insights/#:~:text=AWS%20is%20the%20fastest%20growing%20and%20most%20profitable%20segment.&text=AWS%20services%20continued%20to%20grow,the%20company's%20most%20profitable%20segment.)).

##  AWS SaaS Sales Data

Berikut merupakan dataset original:

1. Row ID: A unique identifier for each transaction.
2. Order ID: A unique identifier for each order.
3. Order Date: The date when the order was placed.
4. Date Key: A numerical representation of the order date (YYYYMMDD).
5. Contact Name: The name of the person who placed the order.
6. Country: The country where the order was placed.
7. City: The city where the order was placed.
8. Region: The region where the order was placed.
9. Subregion: The subregion where the order was placed.
10. Customer: The name of the company that placed the order.
11. Customer ID: A unique identifier for each customer.
12. Industry: The industry the customer belongs to.
13. Segment: The customer segment (SMB, Strategic, Enterprise, etc.).
14. Product: The product was ordered.
15. License: The license key for the product.
16. Sales: The total sales amount for the transaction.
17. Quantity: The total number of items in the transaction.
18. Discount: The discount applied to the transaction.
19. Profit: The profit from the transaction.

### Uji Korelasi 
- Sales vs Discount
- Sales vs Profit
- Order Date (Year) vs Sales

#### Sales vs Discount

```python
from scipy.stats import spearmanr

correlation_value, p_value = spearmanr(df_clean['Sales'], df_clean['Discount'])
print(f"Korelasi Spearman: {correlation_value}, p-value: {p_value}")
```

```python
# conclusion
alpha = 0.05
if p_value < alpha:
    print("Tolak H0: Ada hubungan monotonik yang signifikan antara Sales dan Dsicount.")
else:
    print("Gagal menolak H0: Tidak ada hubungan monotonik yang signifikan antara Sales dan Discount.")
```

#### Sales vs profit

```python

correlation_value, p_value = spearmanr(df_clean['Sales'], df_clean['Profit'])
print(f"Korelasi Spearman: {correlation_value}, p-value: {p_value}")
```

```python
# conclusion
alpha = 0.05
if p_value < alpha:
    print("Tolak H0: Ada hubungan monotonik yang signifikan antara Sales dan Profit.")
else:
    print("Gagal menolak H0: Tidak ada hubungan monotonik yang signifikan antara Sales dan Profit.")
```

#### Order date (Year) vs Sales

```python
df_clean['Order Year'] = df_clean['Order Date'].dt.year
```

```python
salesXyear = df_clean.groupby('Order Year')['Sales'].sum()
```

```python
correlation_dfOY = pd.DataFrame({'Order Year': salesXyear.index, 'Total Sales': salesXyear.values})
```

```python
correlation_value, p_value = spearmanr(correlation_dfOY['Order Year'], correlation_dfOY['Total Sales'])
print(f"Korelasi Spearman: {correlation_value}, p-value: {p_value}")
```

```python
# conclusion
alpha = 0.05
if p_value < alpha:
    print("Tolak H0: Ada hubungan monotonik yang signifikan antara Order Year dan Total Sales.")
else:
    print("Gagal menolak H0: Tidak ada hubungan monotonik yang signifikan antara Order Year dan Total Sales.")
```
